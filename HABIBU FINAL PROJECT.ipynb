# DATA CLEANING

## 5. REMOVING DUPLICATES

```python
# removing duplicates from the whole dataset
df = df.drop_duplicates()
```

```python
# displaying parts where duplicates where removed
df.tail()
```

## 6. DROPPING UNWANTED COLUMNS

```python
# removing columns
df = df.drop(columns = "Ram_Card")
```

```python
# displaying head to confirm removed column
df.head()
```

## 7. REMOVING UNWANTED CHARACTERS IN COLUMNS

```python
# removing unwanted characters from "Day_of_Week" column
df["Brand"] = df["Brand"].str.strip("..._/")
```

```python
# displaying dataset to view removed characters
df.head(10)
```

## REPLACING UNWANTED ELEMENT IN BRAND COLUMN

```python
df["Brand"] = df["Brand"].str.replace("HP…", "HP")
```
